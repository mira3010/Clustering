# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_reducer = PCA(n_components=2)
reduced_belief = pca_reducer.fit_transform(belief)
```

# Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
cluster_distorsion = []
for i in range(1, 10):
    test_sklearn = KMeans(n_clusters=i)
    test_sklearn.fit(reduced_belief)
    cluster_distorsion.append(test_sklearn.inertia_)
plt.figure(figsize=(9, 6))
plt.plot(cluster_distorsion, marker='o', label='Distorsion for some number of clusters')
plt.title("Distorsion between clusters")
plt.xlabel('Number of clusters')
plt.ylabel("Sum of square error")
plt.legend()
plt.show()
```

# Silhoutte Coefficient

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
from matplotlib import cm
```

```python
model = KMeans(
    n_clusters=3,
    init='k-means++')

model = model.fit(reduced_belief)

y = model.predict(reduced_belief)

cluster_labels = np.unique(y)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(reduced_belief, y, metric='euclidean')

y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper),
            c_silhouette_vals,
            height=1.0,
            edgecolor='none',
            color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
silhouette_avg = np.mean(silhouette_vals)

plt.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()

print ("Silhouette coefficient is", silhouette_avg)
y
```

# Kmeans

```python
from sklearn.cluster import KMeans
import collections
```

```python
model = KMeans(
    n_clusters=2,
    init='k-means++',
    # n_init=10,
    # max_iter=300,
    # n_jobs=-1,
    random_state=1)

model = model.fit(reduced_belief)

test = model.predict(reduced_belief)
print(test)
plt.figure(figsize=(10, 6))
plt.scatter(reduced_belief[test== 0,0], reduced_belief[test == 0,1], c='purple', label='Cluster 1')
plt.scatter(reduced_belief[test== 1,0], reduced_belief[test == 1,1], c='green', label='Cluster 2')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 200, c = 'black',  marker='*', label = 'Centroids')
plt.title('Clusters')
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.legend()
plt.grid()
plt.show()

print('K Means Result : ')
print(collections.Counter(test))
```

```python
high= np.count_nonzero(test==1)
low= np.count_nonzero(test==0)
```

```python
percentHigh= high/2407 * 100
print("Cluster 2 is", "%.2f" % percentHigh, "%")
percentLow= low/2407 * 100
print("Cluster 1 is", "%.2f" %  percentLow, "%")
```
