```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("D:/FYP/data/normbelief.csv")

data.head()
```

```python
belief= data[['C_C1_4A', 'C_C1_4B', 'C_C1_4C']]
belief
```

```python
from sklearn.decomposition import PCA
pca_reducer = PCA(n_components=2)
reduced_belief = pca_reducer.fit_transform(belief)
pcadata= pd.DataFrame(reduced_belief,columns= ['pca1','pca2'])
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))  

plt.title("Dendrograms")  

dend = shc.dendrogram(shc.linkage(reduced_belief, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
import collections

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  

y = cluster.fit_predict(reduced_belief)
print (y)

plt.figure(figsize=(10, 7))

plt.scatter(reduced_belief[y==0,0], reduced_belief[y == 0,1], s=100, c='red')
plt.scatter(reduced_belief[y==1,0], reduced_belief[y== 1,1], s=100, c='black')

plt.title('Clusters')
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.grid()
plt.show()

print('Result : ')
print(collections.Counter(y))
```

```python
high= np.count_nonzero(y==1)
low= np.count_nonzero(y==0)

percentHigh= high/2407 * 100
print("Cluster 2 is", "%.2f" % percentHigh, "%")
percentLow= low/2407 * 100
print("Cluster 1 is", "%.2f" %  percentLow, "%")
```
